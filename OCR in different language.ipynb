### 測試Google pytesseract OCR 套件在不同語言下的表現

```python
# 載入套件

import pytesseract
import matplotlib.pyplot as plt
from PIL import Image

pytesseract.pytesseract.tesseract_cmd = r'D:/Tesseract-OCR/tesseract.exe'
dpis = [250,300,350,400]

def open_show_image(file_path,x,y):
    image = Image.open(file_path)
    plt.figure(figsize = (x,y))
    plt.xticks([])
    plt.yticks([])
    plt.imshow(image)
    
def show_result(file_path,lang):
    for dpi in dpis:
        result = str(pytesseract.image_to_string(Image.open(file_path),config="--dpi " + str(dpi),lang=lang)).replace(" ","")
        print("dpi:" + str(dpi))
        print("\nResult:")
        print(result)
        print("="*25)
```

#### 一號選手 50音表

```python
file_path = r"D:\100. Issue\15. SE_pdf\Image 1.jpg"
open_show_image(file_path,8,8)
```

圖中包含英文繁中及日文，試著一次解析看看

```python
show_result(file_path,lang = "eng+chi_tra+jpn")
```

中英日夾雜效果非常的差，只有抓到紅底白字的中文  

#### 二號選手 Power BI登入畫面

```python
file_path = r"D:\100. Issue\15. SE_pdf\Image 2.jpg"
open_show_image(file_path,10,10)
```

這次圖片混合中英

```python
show_result(file_path,"chi_tra+eng")
```

這張圖片當DPI到400時就辨識不出繁體中文了，白底黑字的微軟正黑體繁體中文辨識看起來還算不錯

#### 三號選手 一段日文文章 + 中文翻譯

```python
file_path = r"D:\100. Issue\15. SE_pdf\Image 3.jpg"
open_show_image(file_path,10,10)
```

此張圖片是一段日文文章，每行日文都會夾雜一段繁體中文翻譯

```python
show_result(file_path,"chi_tra+jpn")
```

中日文夾雜的文字容易讓辨識軟體產生錯亂，導致許多辨識錯誤的地方，不過在DPI 250時辨識效果還是不錯

#### 四號選手 日本風量預測

```python
file_path = r"D:\100. Issue\15. SE_pdf\Image 4.jpg"
open_show_image(file_path,10,10)
```

此圖片中包含 日本漢字、日文、日本漢字上的日文注釋，以及圖片中的圖片，圖片中的圖片還有文字

```python
show_result(file_path,"jpn")
```

漢字上的注釋很明顯的也被辨識出來，打亂整個版面  
在辨識上的效果 dpi 300 時是不錯的

#### 五號選手 英文+簡體中文

```python
file_path = r"D:\100. Issue\15. SE_pdf\Image 5.jpg"
open_show_image(file_path,10,10)
```

此張圖片有簡體中文與英文，其中英文還是紅底紅字

```python
show_result(file_path,"chi_sim+eng")
```

DPI 300時有不錯的辨識效果，中文英文都辨識到位，除了少數標點符號被誤判

結論：
1. 對於文字，只要文字與底色的對比度夠高，Google pytesseract可以有機會辨識出來
2. 中(繁體or簡體)英挾雜的圖片，因兩個文字特徵較能明顯判斷，因此一起辨識時效果還算不錯
3. 圖片只有單一日文時，辨識效果不錯，若日文中有漢字且漢字上有注釋的話，則會一起辨識，打亂辨識版面
4. 中文與日文雖也能高度辨識，不過也會造成小部分辨識混亂
5. 尚未找到日文+英文的文章圖片做辨識，猜測英+日的辨識效果不會太好 (因為筆劃都少，特徵難以區分)
6. 做了五張圖片，目前看起來DPI 250 - 300 有很不錯的辨識效果
7. 下次嘗試對圖片做一些增刪修或是改成黑白試試辨識效果
8. 如果能直接用電子文字檔還是比較好的
