# 圖形化列印數字-陣列巢狀迴圈判斷 #
## 說明 ##
請試著輸入一整數數字N(N長度<=10)<br>
若不符合條件，則印出`Input error, please enter again!`<br>

(y/n)對應到　→　該數字符號 / 星號 <br>
根據符號規則列印出對應圖形的數字N<br>

## Hint ##

 ```

1.注意各數字陣列裡的空白皆要印出 

2.可使用陣列的.replace()，將*取代成該數字的符號。

3.使用陣列搭配判斷式和迴圈。

4.此題目為交錯式輸入輸出，操作時會如下：
 ```

````python
Zero = ["  ***  ",
        " *   * ",
        "*     *",
        "*     *",
        "*     *",
        " *   * ",
        "  ***  "]
One = [" * ",
       "** ",
       " * ",
       " * ",
       " * ",
       " * ",
       "***"]

Two = [" *** ",
       "*   *",
       "*  * ",
       "  *  ",
       " *   ",
       "*    ",
       "*****"]

Three = [" *** ",
         "*   *",
         "    *",
         "  ** ",
         "    *",
         "*   *",
         " *** "]

Four = ["   *  ","  **  "," * *  ","*  *  ", "******","   *  ","   *  "]
Five = ["*****", "*    ", "*    ", " *** ", "    *", "*   *", " *** "]
Six = [" *** ", "*    ", "*    ", "**** ", "*   *", "*   *", " *** "]
Seven = ["*****", "    *", "   * ", "  *  ", " *   ", "*    ", "*    "]
Eight = [" *** ", "*   *", "*   *", " *** ", "*   *", "*   *", " *** "]
Nine = [" ****", "*   *", "*   *", " ****", "    *", "    *", "    *"]
````

````python
def ShowTheNumber(N=N,C=C):
    if N >= 0 and C in ("Y","N") and len(str(N)) <= 10:
        arr = [Zero,One,Two,Three,Four,Five,Six,Seven,Eight,Nine]
        for j in range(7):  
            a = ''
            for i in range(len(str(N))):
                if C == "Y":
                    a += arr[int(str(N)[i])][j].replace('*',str(N)[i]) + '  ' 
                else:
                    a += arr[int(str(N)[i])][j] + '  ' 
            print(a,"\t")
    else:
        print("Input error, please enter again!")

        
````

````python
N = int(input())
C = input()
ShowTheNumber(N,C)
````
